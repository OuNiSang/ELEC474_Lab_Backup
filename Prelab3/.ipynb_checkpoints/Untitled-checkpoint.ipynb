```python
import cv2 as cv
import numpy as np
import random
import math
from matplotlib import pyplot as plt
```

```python
global img1Name,img2Name,windowName_1,trackbarName_1,trackbarValRange_1,imgTitle
img1Name = "spiral.jpg"
img2Name = "lines.jpg"
windowName_1 = "Prelab3_GenralEdge_Detection_20119641"

trackbarName_1 = "t Value"
trackbarValRange_1 = [0,100]

imgTitle = np.array((
"Sobel Horizontal Image",
"Sobel Vertical Image",
"Gradient Image"    
))
```

```python
def GeneralEdgeDetection(img,imgSize,t,windowName):
    imgGray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
    # plt.subplot(4,1,3)
    # plt.imshow(cv.cvtColor(imgGray,cv.COLOR_BGR2RGB))
    # plt.title("Original Image")
    imgGray = cv.GaussianBlur(imgGray,(5,5),10)
    # plt.subplot(4,1,4)
    # plt.imshow(cv.cvtColor(imgGray,cv.COLOR_BGR2RGB))
    # plt.title("GaussianBlur Original Image")

    #filtering x and y 
    kernelX = np.array((
    [-1,0,1],
    [-2,0,2],
    [-1,0,1]        
    ),dtype="float64")

    kernely = np.array((
    [-1,-2,-1],
    [0,0,0],
    [1,2,1]        
    ),dtype="float64")

    imgGx = np.absolute(cv.filter2D(imgGray,-1,kernelX))
    imgGy = np.absolute(cv.filter2D(imgGray,-1,kernely))
    
    #imshow




    # imgUnion = cv.addWeighted(imgGx,0.5,imgGy,0.5,0)
    imgUnion = np.absolute(imgGx) + np.absolute(imgGy)



    imgOut = np.zeros((imgSize[0],img1Size[1]))
    for x in range(imgSize[0]):
        for y in range(imgSize[1]):
            if imgUnion[x][y] > t:
                imgOut[x][y] = 255
    cv.imshow(windowName,imgOut)
    
    imgPltSum = []
    imgPltSum.append(imgGx)
    imgPltSum.append(imgGy)
    imgPltSum.append(imgUnion)
    return imgPltSum
```

```python
def PltImg(img,imgDescipt):
    plt.figure(dpi=200)
    plt.figure(figsize=(10,10))

    plt.imshow(cv.cvtColor(img ,cv.COLOR_BGR2RGB))
    plt.title(imgDescipt)
    plt.tight_layout()
```

```python
def TrackbarCallback(*args):
    t_1 = cv.getTrackbarPos(trackbarName_1,windowName_1)
```

```python
def initTrackbar():
    cv.namedWindow(windowName_1, cv.WINDOW_NORMAL)
    cv.resizeWindow(windowName_1,img1Size[1], img1Size[0])
    cv.createTrackbar(trackbarName_1, windowName_1, trackbarValRange_1[0],trackbarValRange_1[1], TrackbarCallback)
```

```python
img1 = cv.imread(img1Name)
img2 = cv.imread(img2Name)
img1Size = [img1.shape[0],img1.shape[1]]
img2Size = [img2.shape[0],img2.shape[1]]

initTrackbar()

global t_1
t_1 = 0

cv.imshow(windowName_1,img1)
while True:
    key = cv.waitKey(1)
    if key == 27: #press ESC to exit the program
        break
    if key == 13:
        t_1 = cv.getTrackbarPos(trackbarName_1,windowName_1)
        imgPltSum = GeneralEdgeDetection(img1,img1Size,t_1,windowName_1)
for i in range(len(imgPltSum)):
    PltImg(imgPltSum[i],imgTitle[i])
cv.destroyAllWindows()
```
