Import

```python
import numpy as np
import cv2 
import random
import math
from matplotlib import pyplot as plt

img = cv2.imread("spiral.jpg")
gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

Smooth image

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
dstimg = cv2.GaussianBlur(img,(3,3),0)
plt.imshow(dstimg)
plt.title("Smoothing image")
```

Get Sobel Horizontal and vertical image 

```python
img = cv2.cvtColor(gray,cv2.COLOR_BGR2RGB)
dstimg = cv2.GaussianBlur(img,(3,3),0)
img = dstimg
G_x = np.array([[-1, 0, 1],[-2, 0, 2],[-1, 0, 1]])
G_y = np.array([[-1, -2, -1],[0, 0, 0],[1, 2, 1]])  
dstx = cv2.filter2D(img,-1,G_x)
dsty = cv2.filter2D(img,-1,G_y)
```

```python
plt.imshow(dstx)
plt.title("Sobel Horizontal Image")
```

```python
plt.imshow(dsty)
plt.title("Sobel Vertical Image")
```

Get gradient image

```python
rows = img.shape[0]
columns = img.shape[1]
Gimg = np.zeros((img.shape[0],img.shape[1],3),np.uint8)
for x in range(rows):
    for y in range(columns):
        Gimg[x][y] = abs(dstx[x][y])+ abs(dsty[x][y])
plt.imshow(Gimg)
plt.title("Gradient Image")
```

Mark the Gradient image

```python
tau = 0
resGimg = np.copy(Gimg)

def tau_val_trackbar(*arg):
    tau = cv2.getTrackbarPos("value of tau", "MarkGrad")


def storeGimg(img):  
    for x in range(rows):
        for y in range(columns):
            img[x][y] = resGimg[x][y]
            
def MarkGrad(img,tau):
    for x in range(rows):
        for y in range(columns):
            for z in range(3):
                if(img[x,y,z] >= tau):
                    img[x,y,z] = 255
    
def active():
    tau = cv2.getTrackbarPos("value of tau", "MarkGrad")
    storeGimg(Gimg)
    MarkGrad(Gimg,tau)
    


cv2.namedWindow("MarkGrad", cv2.WINDOW_NORMAL)
cv2.resizeWindow("MarkGrad",columns, rows)
cv2.createTrackbar("value of tau", "MarkGrad", 0,30, tau_val_trackbar)
while True:
    cv2.imshow("MarkGrad", Gimg)
    key = cv2.waitKey(1)
    if key == 27:   # "esc" to exit
        break
    if key== 13:    #"enter" to active the alogrithm
        active()
cv2.destroyAllWindows()   
```
