```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
import sklearn

print(cv2.__version__)
print(np.__version__)
print(sklearn.__version__)
```

```python
print("Hello World!")
img = cv2.imread("lena.png")
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.title("Lena Color")
```

```python
plt.imshow(gray, cmap = 'gray')
plt.title("Lena Grayscale")
```

```python
print(img.shape)
print(img.size)
print(img.dtype)
```

```python
#------5.2 Flipping the image----------------
# img_hFLip = cv2.flip(img, 0)
# img_gray_hFlip = cv2.flip(gray,0)

imgRow = img.shape[0]
imgCol = img.shape[1]
img_vFlip = np.zeros(img.shape, img.dtype)
img_hFlip = np.zeros(img.shape, img.dtype)
img_gray_vFlip = np.zeros(img.shape, img.dtype)
img_gray_hFlip = np.zeros(img.shape, img.dtype)


for i in range(imgRow):
    for j in range(imgCol):
        img_vFlip[imgRow - 1 - i][j] = img[i][j]
        img_hFlip[i][imgCol - 1 - j] = img[i][j]
        img_gray_vFlip[imgRow - 1 - i][j] = gray[i][j]
        img_gray_hFlip[i][imgCol - 1 - j] = gray[i][j]
        

plt.subplot(2,2,1)
plt.imshow(cv2.cvtColor(img_hFlip, cv2.COLOR_BGR2RGB))
plt.title("Lena color horizontal flip")

plt.subplot(2,2,2)
plt.imshow(cv2.cvtColor(img_vFlip, cv2.COLOR_BGR2RGB))
plt.title("Lena color verticall flip")

plt.subplot(2,2,3)
plt.imshow(img_gray_hFlip, cmap = 'gray')
plt.title("Lena gray horizontal flip")

plt.subplot(2,2,4)
plt.imshow(img_gray_vFlip, cmap = 'gray')
plt.title("Lena gray verticall flip")

plt.tight_layout() 
```

```python
#------------------5.3 Negating iamge pixel-------------

img_neg = np.zeros(img.shape, img.dtype)
img_gray_neg = np.zeros(img.shape, img.dtype)

for i in range(imgRow):
    for j in range(imgCol):
        img_neg[i][j] = 255 - img[i][j]
        img_gray_neg[i][j] = 255 - gray[i][j]
        
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img_neg, cv2.COLOR_BGR2RGB))
plt.title("Lena color negated")

plt.subplot(1,2,2)
plt.imshow(img_gray_neg, cmap = 'gray')
plt.title("Lena gray negated")

plt.tight_layout() 
```

```python
#------------------5.4 Separating Color Channe-------------

img_B = np.zeros(img.shape, img.dtype)
img_G = np.zeros(img.shape, img.dtype)
img_R = np.zeros(img.shape, img.dtype)
img_Merge = np.zeros(img.shape, img.dtype)

for i in range(imgRow):
    for j in range(imgCol):
        img_B[i][j] =  ((img[i][j])[0], 0, 0)
        img_G[i][j] =  (0, (img[i][j])[1], 0)
        img_R[i][j] =  (0, 0, (img[i][j])[2])
        img_Merge[i][j] = img_B[i][j] + img_G[i][j] + img_R[i][j]

plt.subplot(2,3,1)
plt.imshow(cv2.cvtColor(img_B, cv2.COLOR_BGR2RGB))
plt.title("Lena blue image")

plt.subplot(2,3,2)
plt.imshow(cv2.cvtColor(img_G, cv2.COLOR_BGR2RGB))
plt.title("Lena green image")

plt.subplot(2,3,3)
plt.imshow(cv2.cvtColor(img_R, cv2.COLOR_BGR2RGB))
plt.title("Lena red image")

plt.subplot(2,3,5)
plt.imshow(cv2.cvtColor(img_Merge, cv2.COLOR_BGR2RGB))
plt.title("Merged Lena image")

plt.tight_layout() 

```
